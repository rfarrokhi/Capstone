```python
# define a function to choose each country from each dataframe above and then separate years and 
# features to set years as index and features as column variables.
# There is 29 dataframe with 15 numbers for countries.

def convert_func(df, number):
    
    df_T = df.iloc[number].T.squeeze()

    values_features = [df_T[i] for i in range(len(df_T))]

    df_new = pd.DataFrame(df_T)

    index_filter = df_new.filter(like = "num").index
    years = [index_filter[i][:4] for i in range(len(df_T))]
    name_features = index_filter[0][6:]

    df_out = pd.DataFrame(list(zip(years, values_features)), columns = ["years", name_features])

    df_out[:1] = df_out[:1].fillna(df_out.mean())
    df_out[-1:] = df_out[-1:].fillna(df_out.mean())

    df_out = df_out.interpolate(method ='linear', limit_direction ='forward')

    df_out = df_out.set_index("years")

    return df_out
```

### ARIMA model
#### ARIMA(p, d, q)

```python
# Create train-test split.

y_train, y_test = train_test_split(df_IRN["Inflation_GDP_defl"],
                                   test_size = 0.1,
                                   shuffle = False)               # Shuffle = False is IMPORTANT!
```

```python
# Import ARIMA model.

from statsmodels.tsa.arima_model import ARIMA
```

### Manual GridSearch

```python
# Starting AIC, p, and q.

best_aic = 99 * (10 ** 16)
best_p = 0
best_q = 0

# Use nested for loop to iterate over values of p and q.

for p in range(5):
    for q in range(5):
        
        # Insert try and except statements.
        
        try:
            
            # Fitting an ARIMA(p, 1, q) model.
            
            print(f'Attempting to fit ARIMA({p}, 1, {q})')
            
            # Instantiate ARIMA model.
            
            arima = ARIMA(endog = y_train.astype(float).dropna(),       # endog = Y variable
                          order = (p, 1, q))                              # values of p, d, q
            
            # Fit ARIMA model. 
            
            model = arima.fit()

            # Print out AIC for ARIMA(p, 1, q) model.
            
            print(f'The AIC for ARIMA({p}, 1, {q}) is: {model.aic}')

            # Is my current model's AIC better than our best_aic?
            
            if model.aic < best_aic:
                
                # If so, let's overwrite best_aic, best_p, and best_q.
                
                best_aic = model.aic
                best_p = p
                best_q = q

        except:
            pass
print()
print()
print('MODEL FINISHED!')
print(f'Our model that minimizes AIC on the training data is the ARIMA({best_p}, 1, {best_q}).')
print(f'This model has an AIC of {best_aic}.')
```

```python
# Instantiate best model.

model = ARIMA(endog = y_train.astype(float).dropna(),         # endog = Y variable
              order = (0, 1, 2))

# Fit ARIMA model.

arima = model.fit()

# Generate predictions based on test set.

preds = model.predict(params = arima.params,
                      start = y_test.index[0],
                      end = y_test.index[-1])

# Plot data.

plt.figure(figsize = (15, 6))

# Plot training data.

plt.plot(y_train.index, pd.DataFrame(y_train).diff(), color = 'blue')

# Plot testing data.

plt.plot(y_test.index, pd.DataFrame(y_test).diff(), color = 'orange')

# Plot predicted test values.

plt.plot(y_test.index, preds, color = 'green')

plt.title(label = 'Once-Differenced Inflation deflator with ARIMA(0, 1, 2) Predictions', fontsize = 16)
plt.xticks(fontsize = 10, rotation = 75)
plt.show();
```
